## Exercise 1: mean firing rate per brain structure for a given session



```python
import pandas as pd
```

```python
brain_structures = np.unique((Channel&(Probe&'session_id = 715093703')).fetch('brain_structure'))
fr_df = pd.DataFrame(columns=['brain_structure','firing_rate'], 
                     index=range(len(brain_structures)))
fr_df['brain_structure'] = brain_structures
for i,bs in enumerate(brain_structures.astype(str)):
    spiketrain = (SpikeTrain&(Unit&((Channel&(Probe&'session_id = 715093703'))&('brain_structure = "'+bs+'"')))).fetch(format='frame')

    num_units = spiketrain.shape[0]
    firing_rate_area = np.empty((num_units))
    for ii in range(num_units):
        firing_rate_area[ii] = len(spiketrain['spike_ts'].values[ii])/9580.
    fr_df.firing_rate.loc[i] = firing_rate_area.mean()
    
```

```python
plt.plot(fr_df.firing_rate, 'o')
plt.xticks(range(len(fr_df)), fr_df.brain_structure.values);
```

```python
stim_table = (Stimulus()&('stimulus_name = "natural_scenes"')&('session_id = 715093703')).fetch(format='frame')
```

```python
stim_table
```

```python
images = np.unique((Stimulus()&('stimulus_name = "natural_scenes"')&('session_id = 715093703')).fetch('frame'))
```

```python
stim_ids = (Stimulus()&('stimulus_name = "natural_scenes"')&('session_id = 715093703')&('frame="-1"')).fetch('stimulus_presentation_id')
stim_ids
```

```python
responses = np.empty((len(images), 10))
for i,im in enumerate(images):
    stim_ids = (Stimulus()&('stimulus_name = "natural_scenes"')&('session_id = 715093703')&('frame='+str(im))).fetch('stimulus_presentation_id')
    image_response = (TrialSpikeCount()&('session_id = 715093703')&('unit_id in '+str(tuple(unit_ids)))&('stimulus_presentation_id in'+str(tuple(stim_ids)))).fetch(format='frame')
    responses[i,:] = image_response.groupby('unit_id').mean().spike_rate.values
```

```python
responses.argmax(axis=0)-1 #correct for the blank sweep
```

```python
#get trials for single image
#get spikes for single unit, correct for start time
#plot raster/psth whatever is easier
```

```python
for i in range(10):
    image_number = responses.argmax(axis=0)[i]-1
    starts = (Stimulus()&('stimulus_name = "natural_scenes"')&('session_id = 715093703')&('frame='+str(image_number))).fetch('start_time')
    spikes = (SpikeTrain&(Unit&(Channel&(Probe&'session_id = 715093703')))&('unit_id = '+str(unit_ids[i]))).fetch('spike_ts')[0]
    for ii in range(50):
        trial_start = starts[ii]
        rast_x = spikes[np.logical_and(spikes>(trial_start-0.1), spikes<(trial_start+0.35))]-trial_start
        rast_y = np.repeat(ii,len(rast_x))
        plt.plot(rast_x,rast_y, '.', color='k')
        plt.axvspan(0,0.25, color='gray', alpha=0.3)
    plt.title(str(image_number))
    plt.show()
```
